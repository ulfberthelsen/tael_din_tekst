# Introduktion til markdown-funktionen i Jupyter Notebooks

Nedenfor finder I eksempler på de mest almindelig koder, der bruges til markdown-formattering i Jupyter.

* Hvis I dobbeltklikker på feltet kan i se de koder. jeg har brugt.
* Hvis i trykker `SHIFT + ENTER` vises teksten med formateringerne.
* Tre asterisker laver en linje

***

### Overskrifter
Brug `# + mellemrum`  til at lave titler og overskrifter:

    # for titler
    ## for overskrifter
    ### for underoverskrifter
    #### for under-underoverskrifter
***

### Fremhævet skrift:

    Fed: __string__ or **string**
    Kursiv: _string_ or *string*
    
Fed ser **således** ud, og kursiv ser *således* ud.
***


### Linjeskift og indryk
    
    Et linjeskift kan fremtvinges med: <br> 
    Indrykning laves med større4 end-tegnet (>)

Tvunget linjeskift ser <br> således ud

Indrykning ser 
> således ud - indryknigen fortsætter

frem til næste linjeskift
***


### Punktopstillinger
Bullet points kan laves på forskellige måder:

    - med bindestreg
        - brug Tab til at lave underpunkt
       
    * med asterisk 
        * brug Tab til at lave underpunkt
       
Det ser således ud:
- med bindestreg
    - brug Tab til at lave underpunkt
* med asterisk 
    * brug Tab til at lave underpunkt

Du kan oogså bruge tal:

    1. Nummereret punkt
        1. Numereret underpunkt
    1. Du kan bruge 1 hver gang
        1. Jupyter laver rækkefølgen for dig

Det ser således ud:
1. Nummereret punkt
    1. Numereret underpunkt
1. Du kan bruge 1 hver gang
    1. Jupyter laver rækkefølgen for dig
***


### Monospace font
Monospace font laves ved at sætte *accent grave* (\`) på hver side af teksten. `Mono` space bruges ofte til at angive at noget er et fil navn eller et kode-eksempel.

    Fx `print("Hello world!")` 
    
Eksempel: Programmet `print("Hello world!")` er som regel den første linje kode, man lærer.
***

    
### Matematiske symboler
Matematiske symboler placeres mellem to dollar-tegn:

    Fx $\alpha + \omega = \phi$

Det ser således ud: $ \alpha + \omega = \phi $
